```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

#Function to compute the Lorentz curve and Gini Coeficient from the areas under the curves.
def compute_lorenz_curve_and_gini(data):
    # Sort the data in ascending order
    sorted_data = np.sort(data)
    
    # Compute the cumulative distribution function (CDF)
    n = len(sorted_data)
    cdf = np.cumsum(sorted_data) / np.sum(sorted_data)
    
    # Compute the Lorenz curve
    lorenz_curve_x = (np.arange(1, n + 1) / n)*100
    lorenz_curve_y = (np.cumsum(sorted_data) / np.sum(sorted_data))*100
    
    # Compute the area under the Lorenz curve
    lorenz_area = np.trapz(lorenz_curve_y, x=lorenz_curve_x)
    
    # Compute the Gini index
    gini_index = (5000 - lorenz_area) / 5000
    
    return lorenz_curve_x, lorenz_curve_y, gini_index

#Function to compute the Gini Coeficient from the Mean absolute (Equation (6)).
def compute_gini_index_meanabs(data):
    # Sort the data in ascending order
    sorted_data = np.sort(data)
    n = len(sorted_data)
    # Compute the cumulative share of the variable of interest (CDF)
    cdf = np.arange(1, n + 1) / (n + 1)
    # Compute the mean of the dataset
    mu = np.mean(sorted_data)
    # Compute the mean absolute difference
    diff_matrix = np.abs(np.subtract.outer(sorted_data, sorted_data))
    # Compute the Gini index
    gini_index = np.sum(diff_matrix) / (2 * n**2 * mu) 
    return gini_index
```

```python
#Read the data files:
input1000 = np.loadtxt("histogramfitness.dat", skiprows=0, max_rows=1000)
input100 = np.loadtxt("histogramfitness2.dat", skiprows=0, max_rows=100)
input1100 = np.loadtxt("histogramtotal.dat", skiprows=0, max_rows=1100)

# Compute Lorenz curves and Gini indices for each dataset
lorenz_curve_x1, lorenz_curve_y1, gini_index1 = compute_lorenz_curve_and_gini(input1000)
lorenz_curve_x2, lorenz_curve_y2, gini_index2 = compute_lorenz_curve_and_gini(input100)
lorenz_curve_x3, lorenz_curve_y3, gini_index3 = compute_lorenz_curve_and_gini(input1100)

# Plot the Lorenz curves
plt.figure(figsize=(6, 6))

plt.plot(lorenz_curve_x1, lorenz_curve_y1, label="$N_\mathrm{A}=1000$" + "," + " " + "$G=%.3f$" %(gini_index1))
plt.plot(lorenz_curve_x2, lorenz_curve_y2, label="$N_\mathrm{B}=100$" + "," + " " + "$G=%.3f$" %(gini_index2))
plt.plot(lorenz_curve_x3, lorenz_curve_y3, label="$N_\mathrm{T}=1100$" + "," + " " + "$G=%.3f$" %(gini_index3))

plt.plot([0, 100], [0, 100], '--', color='gray', label="Perfect equality")
plt.xlabel("Cumulative percentage of agents", fontsize=16)
plt.ylabel("Cumulative percentage of fitness", fontsize=16)
plt.legend(fontsize="11")
# Increase the size of ticks
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid(True)
plt.savefig('ginilorentz1000100.pdf', bbox_inches='tight')

#Gini index from Equation (6):
print("Gini index (1000):", compute_gini_index_meanabs(input1000))
print("Gini index (100):", compute_gini_index_meanabs(input100))
print("Gini index (1100):", compute_gini_index_meanabs(input1100))
```
