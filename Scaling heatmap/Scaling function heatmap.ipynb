**Heat map of the scaling function for the maximum 
fitness normalized by the sum agent fitnes (Equation 11)**

```python
#Changing to LaTeX font.
import matplotlib
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from numpy import exp,arange

b = np.arange(0, 15, 0.1)
d = np.arange(1, 10000, 50)

B, D = np.meshgrid(b, d)
nu = 1/(1+exp(-B)*(D-1))

contour_levels = np.linspace(0, 1, 11)
plt.contourf(B, D, nu, levels=contour_levels, vmin=0, vmax=1)
cbar = plt.colorbar()
cbar.set_label(r'$F_\mathrm{max}^\mathrm{A}/F_\mathrm{tot}^\mathrm{A}$', fontsize=18)
cbar.ax.tick_params(labelsize=16)
plt.clim(0,1)
plt.xlabel('$\eta$', fontsize=16)
plt.ylabel('$N_\mathrm{A}$', fontsize=16)
# Increase the size of ticks
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
#plt.show()
plt.savefig('scaling.pdf', bbox_inches='tight')
```

**Heat map of the scaling function for the total fitness (Equation 12)**

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from numpy import exp,arange

b = np.arange(0, 15, 0.1)
d = np.arange(1, 10000, 50)

B, D = np.meshgrid(b, d)
nu = (1+exp(-B)*(D-1))/(2+exp(-B)*(10000-1))

contour_levels = np.linspace(0, 1, 11)
plt.contourf(B, D, nu, levels=contour_levels, vmin=0, vmax=1)
cbar = plt.colorbar()
cbar.set_label(r'$F_\mathrm{tot}^\mathrm{A}$', fontsize=18)
cbar.ax.tick_params(labelsize=16)
plt.clim(0,1)
plt.xlabel('$\eta$', fontsize=16)
plt.ylabel('$N_\mathrm{A}$', fontsize=16)
# Increase the size of ticks
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
#plt.show()
plt.savefig('scalingFtot.pdf', bbox_inches='tight')
```
